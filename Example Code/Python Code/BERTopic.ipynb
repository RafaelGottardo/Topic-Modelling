```python
# Import Data

ON_22 = pd.read_csv("Code - Topic Modelling CSDC/data/ON22_newspapers.csv")
```

```python
# Define UMAP and BERTopic model
umap = UMAP(n_neighbors=15,
            n_components=5,
            min_dist=0.0,
            metric='cosine',
            random_state=1998 # Setting the seed to 42 ensures reproducibility
          )

bt_model = bt.BERTopic(language="english", calculate_probabilities=True, umap_model=umap)
```

```python
# Fit Topic Model 

topics, probs = bt_model.fit_transform(ON_22['text'])
```
